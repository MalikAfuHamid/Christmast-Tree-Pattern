# Christmas Tree Pattern



````python
#DAUN ATAS
def Pohon_Natal1(daun_atas): #Outer loop
    
    for i in range (1,daun_atas+1): #Inner loop
        
        #SEGITIGA SIKU-SIKU SEBELAH KIRI
        for j in range (1,daun_atas-i+7):
            print(end= "   ")
        for j in range (i,0,-1):
            print("*", end="  ")
        
        #SEGITIGA SIKU-SIKU SEBELAH KIRI
        for j in range(2,i+1):
            print("*",end="  ")
        print()

#DAUN TENGAH
def Pohon_Natal2(daun_tengah):#Outer loop
    
    for i in range (1,daun_tengah+1): #Inner loop
        
        #SEGITIGA SIKU-SIKU SEBELAH KIRI
        for j in range (1,daun_tengah-i+4):
            print(end= "   ")
        for j in range (i,0,-1):
            print("*", end="  ")
        
        #SEGITIGA SIKU-SIKU SEBELAH KANAN
        for j in range(2,i+1):
            print("*", end="  ")
        print()

#DAUN DASAR
def Pohon_Natal3(daun_dasar): #Outer loop
    
    for i in range (1,daun_dasar+1): #Inner loop
        
        #SEGITIGA SIKU-SIKU SEBELAH KIRI
        for j in range (1,daun_dasar-i+1):
            print(end= "   ")
        for j in range (i,0,-1):
            print("*", end="  ")
        
        #SEGITIGA SIKU-SIKU SEBELAH KIRI
        for j in range(2,i+1):
            print("*",end="  ")
        print()

#BATANG POHON        
def Pohon_Natal4(x):
    for i in range(x):  #Outer loop
        for j in range(4*(x)): #Inner loop
            print('', end= ' ')
        print(' *  *  *')
        

Pohon_Natal1(3)
Pohon_Natal2(6)
Pohon_Natal3(9)
Pohon_Natal4(5)
````

````python
#pembuktian ruang kosong (space)

#DAUN DASAR
def Pohon_Natal3(daun_dasar): #Outer loop
    
    for i in range (1,daun_dasar+1): #Inner loop
        
        #SEGITIGA SIKU-SIKU SEBELAH KIRI
        for j in range (1,daun_dasar-i+1):
            print(end= "   ")
        for j in range (i,0,-1):
            print("*", end="  ")
        
        #SEGITIGA SIKU-SIKU SEBELAH KIRI
        for j in range(2,i+1):
            print("*",end="  ")
        print()

#BATANG POHON        
def Pohon_Natal4(x):
    for i in range(x):  #Outer loop
        for j in range(4*(x)): #Inner loop
            print('', end= 'x')
        print(' *  *  *')

Pohon_Natal3(9)
Pohon_Natal4(5)
````
